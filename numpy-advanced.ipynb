```python
""" 
np.cov() 是 NumPy 库中用于计算协方差矩阵的函数。

rowvar=False 是这个函数的一个关键参数。
默认情况下，np.cov() 假设数据的每一行代表一个变量（特征），
每一列代表一个观测值（样本）。但是，在机器学习和数据分析的习惯中，
通常每一行代表一个样本，每一列代表一个特征。
因此，我们将 rowvar 设置为 False，告诉 np.cov() 函数我们的数据是按列表示特征的。


A positive covariance indicates that the two variables tend to increase or decrease together.   
A negative covariance suggests that when one variable increases, the other tends to decrease, 
and vice versa.   
A covariance close to zero implies that the variables are largely independent of each other.
"""
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 1, 3, 5])

covariance_matrix = np.cov([x, y])
print("Covariance Matrix (Example 1):\n", covariance_matrix)
```

```python
# np.linalg.eig(C)
```

```python
np.linalg.eig(covariance_matrix)
```

```python
evals,PCs = np.linalg.eig(covariance_matrix)
```

```python
evals
```

```python
PCs
```

```python
""" 
np.linalg.eig() 是 NumPy 库中
用于计算方阵的特征值（eigenvalues）和特征向量（eigenvectors）的函数。
我们输入的是协方差矩阵 C。

这行代码会返回两个 NumPy 数组：
evals: 一个包含协方差矩阵 C 的特征值的一维数组。
特征值表示对应特征向量的重要性或方差大小。

PCs: 一个包含特征向量的二维数组。
对于 np.linalg.eig(), 特征向量是按列排列的。
每一列都是一个特征向量，也称为主成分 (Principal Component)。
这些特征向量是相互正交的，并且指向数据中方差最大的方向。
np.linalg.eig() 返回的特征向量已经是单位向量（normalized），
即它们的长度为 1。
"""
```

#### PCs[:,sorter]

含义是：

对所有行，重新排列列的顺序
也就是说：

原来的 PCs 是一个矩阵，每列是一个主成分向量
sorter 是根据特征值降序排列的列索引
你是 在列的维度上进行重排序
🔁 结果：每一列仍是一个主成分方向，顺序是按重要性排序的

```python
# PCs = PCs[sorter]
```

这个操作会：

对行进行重排序
即默认是作用在第一个维度（axis=0）上。

如果 PCs 是一个 (3, 3) 的矩阵，它会将第 0、1、2 行重新排列，而不是列
这会破坏主成分向量的结构（你本来是想调列的顺序）

```python
PCs = np.array([
    [1, 0, 0],  # 第一列是主成分1
    [0, 1, 0],  # 第二列是主成分2
    [0, 0, 1],  # 第三列是主成分3
])
print(PCs[[2,0,1]])
```

```python
print(PCs[:,[2,0,1]])
```

### .copy

```python
# shallow copy
a = np.arange(10)
b = np.arange(10)
c = np.stack((a,b))

index = np.arange(10)
c_shuff = c  
for i in range(2):
    rand = np.random.permutation(index)
    c_shuff[i,:] = c[i,rand]
    print("c:", c)
    print("c_shuff: ",c_shuff)
```

```python
# deep copy

a = np.arange(10)
b = np.arange(10)
c = np.stack((a,b))

index = np.arange(10)
c_shuff = c.copy()
for i in range(2):
    rand = np.random.permutation(index)
    c_shuff[i,:] = c[i,rand]
    print("c:", c)
    print("c_shuff: ",c_shuff)
```
